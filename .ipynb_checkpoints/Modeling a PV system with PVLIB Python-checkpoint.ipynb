 It has been obtained the TMY data for the location selected. This information has been selected from PVGIS database, specifically from the PVGIS-SARAH solar radiation data. PVGIS-SARAH uses the images of the two METEOSAT geostationary satellites (0° and 57°E) covering Europe, Africa and Asia, and that the hourly values are directly calculated from one individual satellite image.

## **-Bonus:** compare the measured DHI from the TMY file with the DHI estimated using a decomposition model of your choice (e.g. Erbs or Reindl) from the measured GHI.

The Erbs model estimates the diffuse fraction DF from global horizontal irradiance through an empirical relationship between DF and the ratio of GHI to extraterrestrial irradiance, Kt. The function uses the diffuse fraction to compute DHI as:
DHI = DF x GHI

```python
# We are going to need the solar position angles from solarposition.
from pvlib import solarposition
times = Location_TMY_df.index
solpos = solarposition.get_solarposition(times, latitude, longitude)
```

```python
solpos.head() #Show some rows
```

```python
ghi = Location_TMY_df['ghi'] #ghi known from TMY data
zenith =  solpos['apparent_zenith'] #apparent zenith from solarposition
datetime_or_doy = solpos.index #time index
```

```python
# Here it has been obtain the values of dhi, dni and ghi from Erbs method.
estimated_radiation = pvlib.irradiance.erbs(ghi, zenith, datetime_or_doy, min_cos_zenith=0.065, max_zenith=87)
```

```python
# Now it will be compared the database diffuse with the diffuse calculated by the decomposition model.

# importing the required module
import numpy as np
import matplotlib.pyplot as plt
  
# x axis values
x = Location_TMY_df['dhi'].resample('M').mean().index.month
# corresponding y axis values
y = estimated_radiation['dhi'].resample('M').mean() #Monthly average of the estimate radiation
y2 = Location_TMY_df['dhi'].resample('M').mean()    #Monthly average of the TMY PVGIS diffuse radiation
  
# plotting the points 
plt.plot(x, y2,'*', label = 'PVGIS')
plt.plot(x, y,'.', label = 'Erbs_Model')
  
# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Diffuse Irradiance (W/m2)')
  
# giving a title to my graph
plt.title('Monthly diffuse mean comparation between PVGIS and Erbs model')
  
# function to show the plot
plt.legend()
plt.show()
```

Here we can see how both methods give similar results, especially in printer and autumn. 

```python
# Now it will be compare the daily results between database diffuse 
# with the diffuse calculated by the decomposition model.

# x axis values
x = list(range(0,365))

# corresponding y axis values
y = estimated_radiation['dhi'].resample('D').mean()  #Daily average of the estimate radiation
y2 = Location_TMY_df['dhi'].resample('D').mean()     #Daily average of the TMY PVGIS diffuse radiation
  
# plotting the points 
plt.plot(x, y2,label = 'PVGIS',alpha=0.6)
plt.plot(x, y,label = 'Erbs_Model', alpha = 0.8)
  
# naming the x axis
plt.xlabel('Days') 

# naming the y axis
plt.ylabel('Diffuse Irradiance (W/m2)')
  
# giving a title to my graph
plt.title('Daily diffuse mean comparation between PVGIS and Erbs model')
  
# function to show the plot
plt.legend()
plt.show()
```

```python
# Function for calculate the iv_parameters needed for plotting the IV Curve.

def plot_iv(pv_module, irr, t_cell, ax, title, label):
    cecparams = pvlib.pvsystem.calcparams_cec(
        effective_irradiance=irr,
        temp_cell=t_cell,
        alpha_sc=pv_module.alpha_sc,
        a_ref=pv_module.a_ref,
        I_L_ref=pv_module.I_L_ref,
        I_o_ref=pv_module.I_o_ref,
        R_sh_ref=pv_module.R_sh_ref,
        R_s=pv_module.R_s,
        Adjust=pv_module.Adjust,
        EgRef=1.121,
        dEgdT=-0.0002677)
    
    IL, I0, Rs, Rsh, nNsVth = cecparams
    
    curve_info = pvlib.pvsystem.singlediode(
    photocurrent=IL,
    saturation_current=I0,
    resistance_series=Rs,
    resistance_shunt=Rsh,
    nNsVth=nNsVth,
    ivcurve_pnts=101,
    method='lambertw')
    
    ax.plot(curve_info['v'], curve_info['i'], label=label)
    ax.plot(curve_info['v_mp'], curve_info['i_mp'], ls='', marker='o', c='black')
    ax.set_title(title)
    ax.set_xlabel('Module voltage [V]')
    ax.set_ylabel('Module current [A]')
    ax.legend()
```

```python
# STC Conditions
E0 = 1000 # W/m2
T0 = 25   # ºC
```

```python
# Plotting the IV Curve for Trina Solar module in STC.
f, ax = plt.subplots()
plot_iv(pv_module=Module_selected, irr=E0, t_cell=T0, ax=ax, title='STC conditions', label='Trina_Solar_TSM_385DE14H_II_')
```

## 5.2. Introduce a module from manufacturer's datasheet

JA Solar JAM72S20 450 W



```python
# Set the electric parameters of the module
v_mp = 41.52 #V
i_mp = 10.84 #A
v_oc = 49.70 #V
i_sc = 11.36 #A
alpha_sc = (i_sc*-0.044)/(100)
beta_voc = (v_oc*-0.272)/(100)
cells_in_series = 48
```

```python
# We can use desoto function to obtain the 5 parameters needed to use the single diode function.
sdm_params = pvlib.ivtools.sdm.fit_desoto(
    v_mp,
    i_mp,
    v_oc,
    i_sc,
    alpha_sc,
    beta_voc,
    cells_in_series,
    EgRef=1.121,
    dEgdT=- 0.0002677,
    temp_ref=25,
    irrad_ref=1000,
    )
```

```python
JA_Solar_params = sdm_params[0]
JA_df = pd.Series(JA_Solar_params)
print(JA_df) #print single diode needed parameters
```

```python
# Function for calculate the iv_parameters needed for plotting the IV Curve with  Desoto Model.

def plot_iv_Soto(pv_module, irr, t_cell, ax, title, label):
    cecparams_soto = pvsystem.calcparams_desoto(
        1000,
        25,
        alpha_sc = JA_df.alpha_sc,
        a_ref = JA_df.a_ref, # Ns*n*Vt
        I_L_ref = JA_df.I_L_ref,
        I_o_ref = JA_df.I_o_ref,
        R_sh_ref = JA_df.R_sh_ref,
        R_s = JA_df.R_s,
        EgRef=1.121,
        dEgdT=- 0.0002677,
        irrad_ref=1000, #G(STC)
        temp_ref=25 #Tcell(STC)
)
    
    IL, I0, Rs, Rsh, nNsVth = cecparams_soto
    
    curve_info_soto = pvlib.pvsystem.singlediode(
    photocurrent=IL,
    saturation_current=I0,
    resistance_series=Rs,
    resistance_shunt=Rsh,
    nNsVth=nNsVth,
    ivcurve_pnts=100,
    method='lambertw')
    
    ax.plot(curve_info_soto['v'], curve_info_soto['i'], label=label)
    ax.plot(curve_info_soto['v_mp'], curve_info_soto['i_mp'], ls='', marker='o', c='black')
    ax.set_title(title)
    ax.set_xlabel('Module voltage [V]')
    ax.set_ylabel('Module current [A]')
    ax.legend()
    return curve_info_soto, cecparams_soto
```

```python
f, ax = plt.subplots()
a,b = plot_iv_Soto(pv_module=JA_df, irr=E0, t_cell=T0, ax=ax, title='STC conditions', label='JA Solar JAM72S20 450 W')
```

# 7. Estimate IAM and Soiling

```python
# First it will be calculated AOI
AOI = pvlib.irradiance.aoi(surface_tilt = surface_tilt, 
                     surface_azimuth = surface_azimuth, 
                     solar_zenith = solar_zenith, 
                     solar_azimuth = solar_azimuth)
```

```python
# IAM loss factor
IAM = pvlib.iam.ashrae(aoi = AOI, 
                 b=0.05)
# Effective_irradiance_1 is the total plane-of-array irradiance less IAM losing factor
effective_irradiance_1 = poa_irradiance['poa_direct']*IAM + poa_irradiance['poa_diffuse'] #effective_irradiance with IAM losses

# Soiling loss factor
# We don't have precipitation data, then we are going to create a rain data serie to watch the result
    
rain = pd.Series(0, index=Location_TMY_df.index)
rain['2022-01-01'] = 1 # mm
rain['2022-02-01'] = 1 # mm
rain['2022-06-01'] = 1 # mm
effective_irradiance_2 = effective_irradiance_1 * (1 - pvlib.soiling.kimber(rainfall=rain)) #effective_irradiance with soiling losses
# Effective_irradiance_2 is the effective_irradiance_1 less this soiling losing supposed

```

```python
# Now it will be compare the total irradiance we have before and after IAM and soiling losses

# x axis values
x = list(range(0,365))

# corresponding y axis values
y = effective_irradiance_2.resample('D').mean()    #Daily mean irradiance with losses
y2 = poa_irradiance['poa_global'].resample('D').mean() #Daily mean irradiance without losses

# plotting the points
plt.plot(x, y,label = 'effective_irradiance')
plt.plot(x, y2,label = 'poa_global')
 
# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Irradiance (W/m2)')
  
# giving a title to my graph
plt.title('Comparative anual mean irradiance with and without losses')
  
# function to show the plot
plt.legend()
plt.show()
```

We can see how losses are considerably larger. This is because  we use for soiling losses only 3 days of rainy days. This result will be misestimated

For this weather data we have precipitable_water all year, then the soling losses will be always zero. This example won't be interesting to observate the application of Kimber model.
```python
rain_fall_df_corrected.plot()  #plot the rainfall data corrected.
(soiling_kimber*100).plot()    #plot the soiling kimber losses in %.
```

# 12. Dimension of the PV Array

```python
# We need the minimun ambient temperature. But this temperature will be important when the module is working.
# This function will calculate the minimum temperature when the irradiance is different from zero, ignoring
# the temperatures of the nigth and sunset.

Min_Temp = 25 
rango = list(range(0,len(Location_TMY_df)))
for i in rango:
    ghi_needed = Location_TMY_df['ghi'][i]
    temp_air_needed = Location_TMY_df['temp_air'][i]
    if (ghi_needed > 0) & (temp_air_needed < Min_Temp):
            Min_Temp =  Location_TMY_df['temp_air'][i] 
print(Min_Temp)
```

```python
# modules per string: max voltage
temp_ref = 25  # °C

# Max Voc module will achieve when the temperature is minimized
max_voc_module = Module_selected.V_oc_ref + Module_selected.beta_oc * (Min_Temp - temp_ref)
modules_per_string = int(inverter_12K['Vdcmax'] / max_voc_module) #total modules per string that inverter can bear

# strings per inverter
power_string = Module_selected.STC * modules_per_string
strings_per_inverter = int(dc_ac_ratio * inverter_12K['Paco'] / power_string) #number string that inverter can bear

dc_capacity = strings_per_inverter * power_string

dc_output = power_module['p_mp'] * modules_per_string * strings_per_inverter

print(f"DC capacity={dc_capacity:,.0f} W.", f"AC capacity (inverter nominal power)={inverter_12K['Paco']:,.0f} W")
print(f"Modules per string={modules_per_string:,.0f}.", f"Strings per inverter={strings_per_inverter:,.0f}")
```

If the site conditions permit this, we can increase the nominal power of the system from 12 kW to 13 kW.

# 9. DC Wiring losses and mismath losses

```python
### Spectral mismatch loss factor
spectral_mismatch_loss = 2/100
dc_output_1 = dc_output * (1-spectral_mismatch_loss)
# dc_output_1 will be the DC power we achieved less spectral mismatch losses.
```

```python
# Para calcular las pérdidas DC vamos a elegir ya un predimensionado del sistema
# como tenemos 12 kW de pottencia nóminal suponemos 40 módulos de 385 W y 5 string de 8
```

```python
print('v_mp:', Module_selected['V_mp_ref'])
print('i_mp:',  Module_selected['I_mp_ref'])
dc_ohmic_percent = 1.5 # typical value of 1.5%
print('DC Ohmic percent:', dc_ohmic_percent ,'%')
```

```python
# Calculates the equivalent resistance of the wires from a percent ohmic loss at STC. 
resistance = pvlib.pvsystem.dc_ohms_from_percent(Module_selected['V_mp_ref'], Module_selected['I_mp_ref'], dc_ohmic_percent, modules_per_string=17, strings=2)   
print('Equivalent resistance of:', resistance, 'Omhs')
```

```python
# Now we calculate the power losses related to DC wiring in watts
power_DC_loss = pvlib.pvsystem.dc_ohmic_losses(resistance, power_module['i_mp'])
```

```python
dc_output_2 = dc_output_1 - power_DC_loss #dc_output with DC wiring losses.
```

# 14. Calculate DC array yield (YA), AC final yield (YF) and reference yield (YR), and then obtain the resulting performance ratio (PR).

### Ya (DC Array Yield)

```python
# We need the nominal power (Pnom) and the energy produced by FV generator (Eg_FV)
Pnom = (Module_selected['STC']*34)/1000 #kWp
Eg_FV = dc_output_2/1000 #kWh
```

```python
Ya = Eg_FV/Pnom
```

```python
Ya.sum()
```

### 14.2 Yf (Final Yield)

```python
# We need the nominal power (Pnom) and the useful energy produced by FV generator(E_FV)
E_FV = ac_output/1000 #kWh
```

```python
Yf = E_FV/Pnom
```

```python
Yf.sum()
```

### Yr (Reference Yield)

```python
# We need the irradiance in the plane of array (Ginc) and the reference irradiance (G_ref)
G_inc = poa_irradiance['poa_global']/1000 #kWh/m2
G_ref = 1 #kW/m2
```

```python
Yr = G_inc/G_ref
```

```python
Yr.sum()
```

### PR (Performance Ratio)

```python
# We can calculate the performance ratio as the division of Yf and Yr
PR = Yf.sum()/Yr.sum()
PR_h = Yf/Yr
PR_h_c = PR_h.replace(-np.inf, 0.0) #Eliminate the infinite values of the nigth results.
print('Total Anual Performance Ratio:', PR)
print('Mean Performance Ratio:', PR_h_c.mean())
```

```python
# We can plot Yf, Ya and Yr to obsrvate the relation

# Set x axis
months = list(range(1,13))
x = months

# Set y axis
y = Ya.resample('M').sum() #Total monthly Dc Array Yield
y2 = Yf.resample('M').sum() #Total monthly Final Yield
y3 = Yr.resample('M').sum() #Total monthly Reference Yield
# y4 = PR_h_c.resample('M').mean()
# corresponding y axis values
plt.plot(x, y,label = 'Ya')
plt.plot(x, y2,label = 'Yf')
plt.plot(x, y3,label = 'Yr')
# plt.plot(x, y4,label = 'PR_h_c')
# plotting the points 
  
# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Horas')
  
# giving a title to my graph
plt.title('Comparative between Yf, Ya and Yr')
  
# function to show the plot
plt.legend()
plt.show()
```

This result fits well with a real power plant. Maybe for following consideration would be interesting to add all the losses related to this system. For this work we only take the most important, in a real case the productivity ratio will decrease if we take into account all the losses.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[],
                     title="PV Sankey Diagram")
sankey = Sankey(ax=ax, unit=None)
sankey.add(flows=[0.89, +0.11 , -0.015, -0.02, -0.965],
           labels=['GHI', 'POA_global', 'IAM', 'Soiling', ''],
           label='Colectors Effective Irradiance',
           orientations=[0, 1, -1, 1, 0])
sankey.add(flows=[0.965, -0.02, -0.0005, -0.03,-0.9145], fc='#37c959',
           label='Nominal Energy',
           labels=['', 'MM', 'Dc wire', 'Inverter', 'Inyected Energy'],
           orientations=[0, 1, -1, 1, 0], prior=0, connect=(4, 0))

diagrams = sankey.finish()
diagrams[-1].patch.set_hatch('/')
plt.legend(loc='lower right')
plt.show()
```

Here we can see how the inpu in "Colectors Effective Irradiance" is lower than the output. This happens because POA_global upgrade the system much more than the losses in this part. Because of this the input flow in the first sankey.add is 0.89 and the output is 0.965.

In "Nominal Energy" we only have losses and the inyected energy will be lower than the input one. Here the flows fo from 0.965 to 0.9145.
